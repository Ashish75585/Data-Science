# Merging, Joining and Concatinating DataFrames
There are 3 main ways of combining DataFrames together. `Merging`, `Joining` and `Concatinating`.

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

## 1. Concatenation
Concatenation basically glues together DataFrames.

<p style="color: green; font-size: 18px">Note:</p>
Dimensions should match along the <code>axis</code> you are concatenating on.</br> 
We can use <code style="background-color: #f0f0f0; padding: 2px 4px; border-radius: 3px;">pd.concat</code> and pass in a <code style="background-color: #f0f0f0; padding: 2px 4px; border-radius: 3px;">list</code> of DataFrames to concatenate together.


```python
pd.concat([df1, df2, df3])
```

The concatenation is done along the default `axis=0` means along the `rows`.

### Concatenating along the Column.

```python
pd.concat([df1, df2, df3], axis=1)
```

## 2. Merging
The `merge` function allows us to `merge` DataFrames together using a similar logic as merging `SQL` tables together.

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

##### The merging is done based on a common `column`.

### 1. Inner Join

```python
pd.merge(left, right, how='inner', on='key')
```

### 2. Outer Join

```python
pd.merge(left, right, how='outer', on='key')
```

### 3. Left Join

```python
pd.merge(left, right, how='left', on='key')
```

### 4. Right Join

```python
pd.merge(left, right, how='right', on='key')
```

#### Example-2: merging based on two keys.

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

### 1. Inner Join

```python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

### 2. Outer Join

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

### 3. Right Join

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

### 4. Left Join

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

## 3. Joining
Joining is a convenient method for combining the columns of two potentially `differently-indexed` DataFrames into a single result DataFrame.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

### 1. Inner Join

```python
left.join(right) # "how" parameter's default value is 'inner"
```

`left.join(right)` means all `rows` of `left` table and common `rows` from `right` table.

### 2. Outer Join

```python
left.join(right, how='outer')
```

In `outer` join, it will take all rows of both tables
