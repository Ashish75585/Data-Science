# Setting colors, linewidths and linetypes

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color='orange') # RGB hex code can be used.
```

### Line width
We can the change the width of the line using `linewidth` parameter. The `default` value is `1`.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color='orange', linewidth=0.8)
```

### `alpha` parameter
This allows us to increase or decrease the transparency of the line.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color='red', linewidth=1, alpha=0.3)
```

### LineStyle
The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# We can also use the shorthand of 'linewidth' which is 'lw'.
axes.plot(x,x+1, color='red', lw=1, alpha=1, linestyle='--')
axes.plot(x,x+2, color='blue', lw=1, alpha=1, linestyle='-.')
axes.plot(x,x+3, color='green', lw=1, alpha=1, linestyle=':')
axes.plot(x,x-1, color='orange', lw=1, alpha=1, linestyle='--')
axes.plot(x,x-2, color='pink', lw=1, alpha=1, linestyle='--')
```

### Markers

```python
x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# We can also use the shorthand of 'linewidth' which is 'lw'.
axes.plot(x,x+1, color='red', lw=1, alpha=1, linestyle='--', marker='o')
axes.plot(x,x+2, color='blue', lw=1, alpha=1, linestyle='-.', marker='+')
axes.plot(x,x+3, color='green', lw=1, alpha=1, linestyle=':', marker='1')
axes.plot(x,x-1, color='orange', lw=1, alpha=1, linestyle='--', marker='s') # 's' means square.
axes.plot(x,x-2, color='pink', lw=1, alpha=1, linestyle='--')
```

#### We can also set the, `markerfacecolor`, `markeredgecolor` and `markeredgewidth`

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# We can also use the shorthand of 'linewidth' which is 'lw'.
axes.plot(x,x+1, color='red', lw=1, alpha=1, linestyle='--', marker='o', markersize=5)
axes.plot(x,x+2, color='blue', lw=1, alpha=1, linestyle='-.', marker='+',  markersize=10)
axes.plot(x,x+3, color='green', lw=1, alpha=1, linestyle=':', marker='1',  markersize=20)
axes.plot(x,x-1, color='orange', lw=1, alpha=1, linestyle='--', marker='s') # 's' means square.
axes.plot(x,x-2, color='pink', lw=1, alpha=1, linestyle='--', marker='s', markeredgecolor='green', markeredgewidth=3, markerfacecolor="yellow")
```

# Plot Range
We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# We can also use the shorthand of 'linewidth' which is 'lw'.
axes.plot(x,y, color='red', lw=1, alpha=1, linestyle='--', marker='o', markersize=5)
# Setting up the x limits.
axes.set_xlim([0,10])
axes.set_xlim([0,20])
```
