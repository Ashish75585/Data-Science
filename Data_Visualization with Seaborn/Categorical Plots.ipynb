# Categorical plots
Categorical Plots are used to plot categorical Data.</br></br>
The mail type of categorical plots are:
* factorplot(deprecated)
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. barplot
A `barplot()` is basically used to `aggregate` the categorical data according to some methods and by default it's the mean.</br>
It can also be understood as a visualization of the group by action. To use this plot we choose a `categorical` column for the `x-axis` and a `numerical` column for the `y-axis`, and we see that it creates a plot taking a `mean` per categorical column.

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

#### Changing the aggregate function.

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)  # np.std --> standard deviation
```

## 2. countplot
`seaborn.countplot()` method is used to Show the counts of observations in each categorical bin using bars.

```python
sns.countplot(x='sex', data=tips) # default value on axis will be the counts.
```

## boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
# Showing the distribution of total bill per day.
# 'day' is a categorical data.
# 'total_bill' is a numerical data.
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# splitting the chart based on another categorical column 'smoker'
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

### Violinplot
Combine a box plot with a kernel density plot to show both summary statistics and the density of the data. The width of the violin's density curve indicates the frequency of data points in each region. 


```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Key Differences Between `Box` and `Violin` Plot

* **Information Displayed:**
    * Box plots only show summary statistics, while violin plots show both summary statistics and the density of the data.
* **Data Distribution:**
    * Violin plots can show the shape of the data distribution, whereas box plots cannot.
* **Outliers:**
    * Violin plots show the entire range of the data, including outliers, while box plots typically only show outliers that fall within 1.5 times the interquartile range.
* **Data Size:**
    * Violin plots can be useful for visualizing small sample sizes, while box plots are suitable for any size dataset.
* **Data Normality:**
    * Violin plots work well with data that is not normally distributed, while box plots may be less effective.
    * Violin plots can be particularly useful when comparing the distributions of multiple groups, as the peaks, valleys, and tails of each group's density curve can be compared.
compared. 

## Stripplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

## Swarmplot
The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

### Combining Categorical plots

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black', size=3)
```
