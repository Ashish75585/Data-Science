# Plotly and Cufflinks
* `Plotly` is an interactive visulization library
* `Cufflinks` connects plotly with pandas

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import cufflinks as cf
```

```python
from plotly import __version__
```

```python
print(__version__)
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# Connecting JavaScript with notebook.
init_notebook_mode(connected=True)
# plotly connects pandas and python with an interactive JS library.
```

```python
cf.go_offline() # this will allows cufflinks to use offline.
```

### Getting the data.

```python
df1 = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df1.head()
```

```python
df2 = pd.DataFrame({'Category':['A', 'B', 'C'], 'Values':[32, 43, 50]})
```

```python
df2.head()
```

### Now, we can use the plots directly.

```python
df1.plot()
```

```python
df1.iplot() # this one is more interactive.
```

## Scatter plot

```python
df1.iplot(kind='scatter', x='A', y='B')
```

WE HAVE TO SPECIFY THE `mode`.

```python
df1.iplot(kind='scatter', x='A', y='B', mode='markers')
```

## Bar plot

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
# Using aggregate function with bar plot.
df1.sum().iplot(kind='bar')
```

## Box plot

```python
df1.iplot(kind='box')
```

## 3D surface plot

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y': [10,20,30,40,50], 'z': [5, 4, 3, 2, 1]})
```

```python
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface', colorscale='rdylbu') # changing the color
```

## histogram

```python
df1['A'].iplot(kind='hist')
```

```python
df1.iplot(kind='hist', bins=25)
```

## Spread

```python
df3[['x', 'y']].iplot(kind='spread')
```

## bubble plot

```python
df1.iplot(kind='bubble', x='A', y='B', size='C') # size is based on another column.
```

## scatter_matrix()
Similar to seaborn's pair plot.

```python
df1.scatter_matrix() # data should be numerical.
```
