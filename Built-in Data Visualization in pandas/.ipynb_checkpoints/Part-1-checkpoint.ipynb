# Pandas built-in Data Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

## Data
There are some fake data csv files you can read in as dataframes.

```python
df1 = pd.read_csv('df1', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# We can directly use the hist() on any column.

df1['A'].hist()
```

```python
# assinging the number of bins
df1['A'].hist(bins=30)
```

### Another way of doing the same as above.

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist()
```

# Plot Types
There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___


### Area

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

### Barplots

```python
df2.plot.bar()
```

### Lineplots

```python
df1.plot.line(y='B') # if 'x' is not specified, then it will use the index of the dataframe by default.
```

```python
df1.plot.line(y='B', figsize=(12,3))
```

### Scatterplot

```python
df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c='C', alpha=0.6) # color based of column 'C'
```

### Boxplots

```python
df2.plot.box()
```

### Hexagonal Bin plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a', y='b')
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
```

### Kernel Density Estimation plot (KDE)

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
